```python
#открывает файл с текстом
#делает частотный список слов и отрезает все, у которых частотность < 3
#по запросу выдает пользователю случайное слово заданной длины на заданную букву

#открывает файл, записывает слова в словарь d
def file(filename):
    symbols = '123456789.,/!?|\@#$%^&*():"{}[];_-=+<>№'
    d = []    
    with open(filename, encoding='utf-8') as f:
        for line in f:
            words = line.split().lower()
            for word in words:
                d.append(word.strip(symbols))
    return d
    
#частотный список слов, длиннее 3

def freq(d):
    import collections
    a = {}    
    counter = collections.Counter(d)
    for key, value in counter.items():
        b = {}
        if value >= 3:
            b = {key: value}
            a.update(b)
    print(a)
    return a
            
#ввод данных, выдача

def inpt(a): 
    import random
    length = input(int())
    letter = input("Letter: ")
    c = {}    
    for key, value in a.items():
        e = {}
        if value == length and key[0] == letter:
            e = {key, value}
            c.update(e)
    print(c)
    need = random.choice(c.keys())
    print(need)
    
    
def main():
    inpt(freq(file('text.txt')))
    
if __name__ == '__main__':
    main()
    

    
    
    
                
            
    
```

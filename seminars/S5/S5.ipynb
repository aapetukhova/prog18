```python
import urllib.request
from bs4 import BeautifulSoup

link = 'https://habr.com/'
result = urllib.request.urlopen(link)
myfile = result.read()
soup = BeautifulSoup(myfile, 'html.parser')
# only text
# print(soup.get_text())
# a pretty version - just the code
# print(soup.prettify()) 
# goes through the links
# for link in soup.find_all('a'):
# #     extracting text from the href attribute
#     print(link.get('href'))
# for every post=article class post prints out its header=a and text=div
for post in soup.find_all('article', {'class': 'post'}):
    print(post.find('a', {'class': 'post__title_link'}).get_text())
    print(post.find('div', {'class': 'post__text'}).get_text())
    print('-'*64)
```

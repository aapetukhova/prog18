```python
# для случайного выбора слова
import random

# для уже использованных букв
used_letters = []

# слово-запоминальщик
word = []

# картинка:
pic = {1: ['_'], 
       2: ['__'], 
       3: ['___', '   |'], 
       4: ['___', '   |', '   O'], 
       5: ['___', '   |', '   O', '  /'], 
       6: ['___', '   |', '   O', '  /|'], 
       7: ['___', '   |', '   O', '  /|\\'], 
       8: ['___', '   |', '   O', '  /|\\', '  /'], 
       9: ['___', '   |', '   O', '  /|\\', '  / \\']}
# ___
#    |
#    O
#   /|\
#   / \

#чтение файлов со словами
def file(filename):
    with open(filename, encoding='utf-8') as f:
        words = []
        for line in f:
            words.append(line.strip('\n'))    
    return words

# выбор темы    
def themes():
    print('1 - Динозавры')
    print('2 - Звезды')
    print('3 - Фараоны Древнего Египта')
    theme = input('Выберите тему цифрой: ')
    return theme

# рандомайзер слов темы
def random(theme):
    import random
    a = []
    if theme == '1':
        a = file('dino.txt')
    elif theme == '2':
        a = file('stars.txt')
    elif theme == '3':
        a = file('skull.txt')
    puzzle = random.choice(a)
    return puzzle

# печать слова
def guess_out(letters):
    print('Слово: {0}'.format(' '.join(letters)))
    
# функция на заданное число оставшихся попыток выдает текст - такой буквы нет - количество оставшихся попыток и рисует человечка         
def hangman(number_of_mistakes):
    print('Такой буквы нет!!!')
    attempts_left = 9 - number_of_mistakes
    if attempts_left >= 5:
        print('У Вас осталось', attempts_left, 'попыток')
    if attempts_left > 1 and attempts_left < 5:
        print('У Вас осталось', attempts_left, 'попытки')
    if attempts_left == 1:
        print('У Вас осталась', attempts_left, 'попытка')    
    for i in pic[number_of_mistakes]:
        print(i)
    if attempts_left == 0:
        print('Вы проиграли!!!')  
    
# игра
def game(puzzle):
    mistakes = 0
    print('У Вас 9 попыток, чтобы отгадать слово')
    #загаданное слово - список, элементы - буквы, в зависимости от того, какие буквы отгаданы, а какие - нет, черточки будут меняться на буквы
    for c in puzzle:
        word.append('_')
#     информация о слове
    print('В слове', len(puzzle), 'букв')
    guess_out(word)
#   если попытки еще есть
    while mistakes < 9:
        guess = input('Введите букву: ').lower()
#       проверка на букву
        if not guess.isalpha():
            print('Это не буква!!!')
#       если букву уже вводили
        elif guess in used_letters:
            print('Вы уже вводили эту букву!!!')
#       буква не была использована
        else:
            used_letters.append(guess)
            if guess in puzzle:
                print('Угадали!')
#               замена черточки на букву
                for i in range(0, len(puzzle)):
                    if puzzle[i] == guess:
                        word[i] = guess
                guess_out(word)
#               если слово отгадано полностью
                if not '_' in word:
                    print('Победа!')
                    break
            else:
                mistakes += 1
                hangman(mistakes) 

#вызов функции
game(random(themes()))      
```
